```python
import json
import requests
```

```python
#json.loads('{"clave":"valor"}')
# ticket mercado publico= 81D9D50E-E3EA-41F5-A456-0AD2B42E6D80
url2 = "http://api.mercadopublico.cl/servicios/v1/publico/licitaciones.xlm?fecha=06102020&estado=todos&ticket=81D9D50E-E3EA-41F5-A456-0AD2B42E6D80 "

respuesta2=requests.request("GET", url2)
print(respuesta2)
```

```python
print(respuesta2.headers)
```

```python
respuesta2.text
```

```python
respuestajson2 = json.loads(respuesta2.text)
```

```python
type(respuestajson2)
```

```python
print(respuestajson2)
```

```python
licitaciones = respuestajson2["Listado"]
```

```python
licitaciones
```

```python
import pandas as pd
```

```python
df_licitaciones_todos = pd.DataFrame(licitaciones, columns=['Nombre','CodigoEstado','FechaCierre','CodigoExterno'])
```

```python
df_licitaciones_todos
```

```python
cantidad = df_licitaciones_todos["Nombre"].value_counts()
```
